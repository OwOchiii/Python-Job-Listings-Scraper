# RealPython Fake Jobs Data EDA
Let's explore the dataset that our scraper collected.

```python
import pandas as pd
import matplotlib.pyplot as plt
from pipeline import process_and_cluster

# Load and cluster data
df, X = process_and_cluster('jobs.csv', n_clusters=5)
```

### Cluster Distribution

```python
if df is not None:
    df['Cluster'].value_counts().plot(kind='bar', title='Count of Jobs per Cluster')
    plt.xlabel('Cluster ID')
    plt.ylabel('Number of Jobs')
    plt.show()
```

### Inspecting Clusters
We can look at top job titles for a specific cluster.
